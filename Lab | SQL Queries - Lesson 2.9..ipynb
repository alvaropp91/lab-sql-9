```python
# 1. How many distinct (different) actors' last names are there?
query1 = 'SELECT count(DISTINCT last_name) \
FROM sakila.actor'
print(query1)
data1 = pd.read_sql_query(query1, engine)
data1.head()
```

```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query2 = '''
SELECT *,
CASE 
WHEN DAYOFWEEK(rental_date)= 1 OR DAYOFWEEK (rental_date) = 7 THEN "Weekend"  
ELSE "Weekday" 
END AS day_type 
FROM sakila.rental
'''
data2 = pd.read_sql_query(query2, engine)
data2
```

```python
# 3. Get all films with ARMAGEDDON in the title
query3 = '''SELECT title 
FROM sakila.film 
WHERE title LIKE "%%ARMAGEDDON%%"'''
data3 = pd.read_sql_query(query3, engine)
data3
```

```python
# 4.Get 10 the longest films
query4 = '''SELECT title, length
FROM sakila.film
WHERE (length > 0)
ORDER BY length
LIMIT 10'''
data4 = pd.read_sql_query(query4, engine)
data4
```

```python
# 5.How many films include Behind the Scenes content?

query5 = '''SELECT count(title)
FROM sakila.film
WHERE special_features LIKE "%%Behind the Scenes%%"'''
data5 = pd.read_sql_query(query5, engine)
data5
```

```python
# 6.Which kind of movies (rating) have a mean duration of more than two hours?
query6 = '''SELECT film.rating
FROM sakila.film
GROUP BY film.rating
HAVING AVG(length) > 120'''
data6 = pd.read_sql_query(query6, engine)
data6
```

```python
# 7.Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

query7 = '''SELECT film.title, film.length
FROM sakila.film
WHERE length IS NOT NULL AND length > 0
ORDER BY length'''
data7 = pd.read_sql_query(query7, engine)
data7
```
